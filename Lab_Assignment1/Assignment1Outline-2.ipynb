```python
# step 3
# define the number of observation n
# (1 mark)
n = len(l) #

# define the observation weight matrix P for LSE as identity matrix
# identity weight matrix simply means the observations will be unweighted in the following LSE
# (1 mark)
P = np.eye(n) #creates an n  n identity matrix. Least Squares Estimation (LSE) means all observations are treated equally

# define the range of function input(i.e., constants a), from start to end; and the numerical array of a
start = -1* np.pi
end = np.pi
a = np.linspace(start, end, n)
```

```python
# step 5
# define a curve fitting function named fit_polynomial_curve
# the function should implement the linear least square estimation
# the function should include all important intermediate steps in LSE
# the input and output of the funciton are described in the comment below in red.
# please keep code clean and well-commented
# (7 marks)

def fit_polynomial_curve(a: np.ndarray, l: np.ndarray, P: np.ndarray, deg: int) -> tuple:
    '''
    Args:
        a:              function input (constants)
        l:              noisy observations
        P:              observation weight matrix
        deg:            degree of polynomial to fit to data
    Returns:
        x:              estimated polynomial coefficients
        cov_x:          uncertainty in estimated coefficients
        apv:            a posterier variance factor (goodness of fit)
        v:              observation residuals
    '''

# define a curve fitting function named fit_polynomial_curve
def fit_polynomial_curve(a, l, P, deg):

    # Construct the design matrix that Columns go from lower powers to higher powers
    A = np.vander(a, N=deg+1, increasing=True)

    # Compute normal equation
    # Multiply the transpose of the design matrix A by the weight matrix P
    # A.T is the transpose of the design matrix A.
    AT_P = A.T @ P
    # Compute normal matrix(N)
    N = AT_P @ A

    #Compute estimated polynomial coefficients
    #np.linalg:Linear Algebra module of NumPy that invert a matrix
    x = np.linalg.inv(N) @ (AT_P @ l)

    # Compute Residuals
    v = l - A @ x

    # Compute variance factor
    dof = len(l) - (deg+1)
    apv = (v.T @ P @ v) / dof

    # Compute Covariance matrix of coefficients
    cov_x = apv * np.linalg.inv(N)

    # Return outputs
    return (x, cov_x, apv, v)
```

```python
# step 6
# call the fit_polynomial_curve function and fetch the results
# use polynomial degrees of: 2, 4, 6
# (1 marks)

# fit polynomial of degree 2
x_2, cov_x_2, apv_2, v_2 = fit_polynomial_curve(a, l, P, deg=2)

# fit polynomial of degree 4
x_4, cov_x_4, apv_4, v_4 = fit_polynomial_curve(a, l, P, deg=4)

# fit polynomial of degree 6
x_6, cov_x_6, apv_6, v_6 = fit_polynomial_curve(a, l, P, deg=6)

#Print the results
print("Degree 2 coefficients:", x_2)
print("Degree 4 coefficients:", x_4)
print("Degree 6 coefficients:", x_6)
```

```python
# step 7
# use the estimated coefficients x and constants array a, construct polynomial functions for each degree
# hint: use polyval function from numpy library (check input argument carefully)
# (1 marks)

# degree 2 fit
quadratic_uw = np.polyval(x_2[::-1], a)

# degree 4 fit
quartic_uw = np.polyval(x_4[::-1], a)

# degree 6 fit
hexic_uw = np.polyval(x_6[::-1], a)
```
